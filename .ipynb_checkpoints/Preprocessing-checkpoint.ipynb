## Preprocessing Features

#### Establish order of ordinal columns

```python
df.columns
```

```python
# identify ordinal columns and define order of the features
# only ordinal column is emp_length
emp_length_list = ['< 1 year',
                   '1 year',
                   '2 years',
                   '3 years',
                   '4 years',
                   '5 years',
                   '6 years',
                   '7 years',
                   '8 years',
                   '9 years',
                   '10+ years']
```

```python
# convert to type category
df['emp_length'] = df['emp_length'].astype('category')
```

```python
# get all nominal columns
nominal_cols = list(df.select_dtypes(include='object').columns)

# get all ordinal columns
ordinal_cols = list(df.select_dtypes(include='category').columns)

# get all numeric columns
numeric_cols = list(df.select_dtypes(include= ['int', 'float']).columns)
```

```python
# get counts of cols
print(len(nominal_cols), "Nominal Columns:", nominal_cols)
print("")
print(len(ordinal_cols), "Ordinal Columns:", ordinal_cols)
print("")
print(len(numeric_cols), "Numeric Columns:", numeric_cols)
```

### Building pipeline transformer

```python
# define a nominal transformation pipeline that OHE the cats
nominal_pipeline = Pipeline([('onehotenc', OneHotEncoder(sparse = False))])

# define an ordinal transformation pipeline that ordinal encodes the cats
ordinal_pipeline = Pipeline([('ordinalenc', OrdinalEncoder(categories = [emp_length_list]))])

# define numeric transformation pipeline that scales the numbers
numeric_pipeline = Pipeline([('numnorm', StandardScaler())]) 
```

```python
# construct column transformer for the selected columns with pipelines
ct = ColumnTransformer(transformers = [("nominalpipe", nominal_pipeline, nominal_cols),
                                       ("ordinalpipe", ordinal_pipeline, ordinal_cols),
                                       ("numericpipe", numeric_pipeline, numeric_cols)])
```

```python
# save the column transformer
dump(ct, open('pkl/column_transformer.pkl', 'wb'))
```

#### Reorder dataframe for simplicity of feature name retrieval

```python
# reorder the df with nominal first, ordinal second, and remaining numeric last
ordered_cols = nominal_cols + ordinal_cols + numeric_cols
df = df[ordered_cols]
```

```python
df.head()
```

## Finding N-Clusters

```python
from sklearn.cluster import MiniBatchKMeans
from sklearn.metrics import silhouette_score
```

### Determining Best Number of Clusters to Select

```python
# stores predicted clustering results
results=[]
kmeans_predictions = []

# iterates for a clustering range of k = 2 to 20
for k in range(2, 20):
    kmeans = MiniBatchKMeans(random_state = 42, 
                             n_clusters=k).fit(sample_df)
    predict = kmeans.predict(sample_df)
    result = {
        'k':k,
        'inertia': kmeans.inertia_,                      # get inertia score for each iteration of k
        'silhouette': silhouette_score(sample_df, predict),  # get silhouette score
    }
    results.append(result)
    kmeans_predictions.append(predict)
    print(result, end='\r')
    
cluster_results = pd.DataFrame(results)
```

### Save Cluster Results and Define Cluster Size

```python
cluster_results.to_csv('data/cluster_results')
```

```python
cluster_results = pd.read_csv('data/cluster_results', index_col= 0)

clusters = 2
```

```python
cluster_results
```

#### Plot Inertia

```python
fig, ax = plt.subplots()
ax.set_title('Inertia vs. Number of Clusters (2 - 20)')
cluster_results.inertia.plot(ax=ax)
ax.vlines(x=clusters, ymin=100000, ymax=1000000, colors='red', linestyles='dotted')
plt.show()
```

#### Plot Silohouette Score

```python
fig, ax = plt.subplots()
ax.set_title('Silhouette Score vs. Number of Clusters (2 - 20)')
cluster_results.silhouette.plot()
ax.vlines(x=clusters, ymin=-.1, ymax=.6, colors='red', linestyles='dotted')
plt.show()
```

Based on these plots to find the optimum of clusters, we can estimate that the number of clusters is approximately 3 based on the elbow plot above.

## Reverse OHE

In order to understand the label assignments better, I'll need to reverse OHE the preprocessed dataframe to get it back to its original form. This will allow for easier interpretation with data visualization. 

Thus, I'll need to format `df_clustered` to look like the original column format in `df` as shown below:

```python
df.head(2)
```

```python
# pp with cluster labels
df_clustered.head(2)
```

```python
# assigns cluster labels
cluster_df = pd.DataFrame(df_clustered['cluster_label']) 

# assigns term
term_df = pd.DataFrame(df_clustered.loc[:,'term_ 36 months':'term_ 60 months'].idxmax(axis = 1)).rename(columns = {0:'term'})

# assigns home ownership
ownership_df = pd.DataFrame(df_clustered.loc[:,'home_ownership_ANY':'home_ownership_RENT'].idxmax(axis = 1)).rename(columns = {0:'home_ownership'})

# assigns loan status
loan_df = pd.DataFrame(df_clustered.loc[:,'loan_status_Charged Off' :'loan_status_Late (31-120 days)'].idxmax(axis = 1)).rename(columns = {0:'loan_status'})

# assigns purpose
purpose_df = pd.DataFrame(df_clustered.loc[:,'purpose_car':'purpose_wedding'].idxmax(axis = 1)).rename(columns = {0:'purpose'})

# assigns state
state_df = pd.DataFrame(df_clustered.loc[:,'addr_state_AK':'addr_state_WY'].idxmax(axis = 1)).rename(columns = {0:'addr_state'})

# assigns app type
app_type_df = pd.DataFrame(df_clustered.loc[:,'application_type_Individual': 'application_type_Joint App'].idxmax(axis = 1)).rename(columns = {0:'application_type'})
                                                                                                           
                                                                                              
# assigns numerical values
numerical_values = df_clustered.loc[:,'emp_length':'total_bc_limit']


# concat all the dfs
reverse_ohe_cluster_data = pd.concat([cluster_df, term_df, ownership_df, loan_df,
                                purpose_df, state_df, app_type_df, numerical_values], axis = 1)

reverse_ohe_cluster_data 
```
