```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# display all columns
pd.set_option('display.max_columns', None)
```

## Data Import & Concat

```python
# read in 2016
q1_16 = pd.read_csv('data/2016Q1.csv.gz')
q2_16 = pd.read_csv('data/2016Q2.csv.gz')
q3_16 = pd.read_csv('data/2016Q3.csv.gz')
q4_16 = pd.read_csv('data/2016Q4.csv.gz')

# read in 2017
q1_17 = pd.read_csv('data/2017Q1.csv.gz')
q2_17 = pd.read_csv('data/2017Q2.csv.gz')
q3_17 = pd.read_csv('data/2017Q3.csv.gz')
q4_17 = pd.read_csv('data/2017Q4.csv.gz')

# combine all the dfs
df = pd.concat([q1_16, q2_16, q3_16, q4_16, 
               q1_17, q2_17, q3_17, q4_17]).reset_index(drop = True)
```

### Handling Missing Values

```python
# missing values percentage totals
(df.isna().mean() * 100).round(2)
```

### Imputing with 0s
Based on the data definitions, I will impute the missing values for the following columns with 0s since these are related to how many months a borrower has been delinquent on payments. Generally, we can assume that this value is 0 as most people tend to pay on time. 

I'll also impute the `emp_length` with the assumption of `< 1 year`. 

```python
# list of columns we can impute with 0s
cols_impute_zeros = ['mths_since_last_delinq', 'mths_since_recent_bc', 'mths_since_recent_bc_dlq',
                     'mths_since_recent_revol_delinq']

# Fill missing values in the specified columns with zeros
df[cols_impute_zeros] = df[cols_impute_zeros].fillna(0)
```

```python
df['emp_length'] = df['emp_length'].fillna('< 1 year')
```

### Imputing Joint App Borrowers?

Let's check the last 4 features involving joint borrowers as this contains almost 420k (96%) missing values. Is there a way to impute these values instead of filling them with just 0s or with the average value?

One way of checking is by filtering the current df for Joint App borrowers and understanding the data better.

```python
# set condition where app type is joint app borrowers
joint_app = df['application_type'].str.contains('Joint App')

# filter df for joint app borrowers
joint_app_df = df[joint_app]
joint_app_df[['revol_bal_joint', 'sec_app_fico_range_high', 'sec_app_earliest_cr_line', 'sec_app_inq_last_6mths']]
```

```python
# how many missing values amongst joint app?
joint_app_df[joint_app_df['revol_bal_joint'].isnull()]
```

So based on the above dataframe filtering, we can conclude the following:
- Joint App borrowers still contain missing values for the 4 last columns. It was assumed that these values would not contain missing values for these columns since these are features related to Joint App borrowers. 
- Thus, these columns likely wont tell us much if we try to impute about 96% of values with median or average values. I will drop these columns moving forward. 

Note: I will also drop all other columns related to joint applications such as `annual_inc_joint`.

#### Dropping `Joint Application` Features

```python
# define joint_app cols
joint_app_cols = ['annual_inc_joint','dti_joint', 'revol_bal_joint',
                  'sec_app_fico_range_high', 'sec_app_earliest_cr_line', 'sec_app_inq_last_6mths']
```

```python
# drop joint_ap cols
df = df.drop(columns = joint_app_cols)
```

```python
len(df['emp_title'].unique())
```

### Dropping Employment Title

For future data modeling simplicity, we'll also drop some additional features below:

- `id`: Contains non-useful info
- `emp_title`: Contains over 120k unique values that will require encoding. This will complicate the modeling performance. 
- `il_util`: Contains a calculated ratio field and does not neccessarily need to be repeated as a feature. 

```python
# drop cols
df = df.drop(columns = ['id', 'emp_title', 'il_util'])
```

### Imputing with Median Values
For all other missing values, we can impute the remaining missing values with the median values of each feature.

```python
df = df.fillna(df.median())
```
